# 5. Model Building and Evaluation

```python
# Model building where am trying to predict the tommorrow's temperature
weather1
```

```python
# Predicting tommorrow's tmax(Maximum Temperature)by first creating a "target" column using the shift() method. The shift()method keeps the same index while pulling the values of the next row back.
weather1["target"] = weather1.shift(-1)["tmax"]
```

```python
# Applying Machine Learning model by first importing Ridge Regression Model which works similarly to linear regression  except it penalizes its coefficients to account for multi corlinearity and that is why I never accounted for corlinearity because Ridge regression factors in multi-corlinearity. 
# Importing Ridge from Scikit-Learn
# Lambda function is used but for this project, I have renamed it alpha
from sklearn.linear_model import Ridge
rr = Ridge(alpha=.1)
```

```python
# Creating a list of predictor columns except the columns listed in the isin()function
predictors = weather1.columns[~weather1.columns.isin(["target","name","station"])]
```

```python
# Calling predictors to check the list of predictor columns
predictors
```

```python
# The weather1 dataset when viewed you realized that tommorrow's data is linked to today's data for example the temperatures on January 2nd 1970 is closely linked to that of January 1st 1970 thus making the dataset to be a time series data thus future data cannot be used to predict past data. Cross-validation is always used in estimating the error of a machine learning model except with time series data for future data cannot be used to predict the past. For time series data it is suitable to use backtesting technique or time series cross-validation to account for the time series nature of the data.def backtest(weather1, model, predictors, start=3650, step=90):
def backtest(weather1, model, predictors, start=3650, step=90): # Generating predictions foe 90 days
    all_predictions = []# List of all predictions
    
    for i in range(start, weather1.shape[0], step):
        train = weather1.iloc[:i,:] # Creating a training set
        test = weather1.iloc[i:(i+step),:] # Creating a test set
    
        model.fit(train[predictors], train["target"])# Fitting the model
        
        preds = model.predict(test[predictors]) # Generating the predictions
        
        preds = pd.Series(preds, index=test.index) # Converting the predictions into a pandas series to change the predictions from the default numpy arrays
        
        combined = pd.concat([test["target"], preds], axis=1) # Concatinating real test data with the prediction using pandas concat() function.
        
        combined.columns = ["actual", "prediction"] # Naming the columns
        
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs() # Creating a column diff which is the prediction  minus the actual while taking the absolute value using abs() function
        
        all_predictions.append(combined) # Combining all_predictions with the combined dataframes using append() function
    return pd.concat(all_predictions) #  Concatination of all predictions using concat() function
     
```

```python
# Generating predictions by calling backtest() while parsing in weather1 dataset, RIDGE REGRESSION AND THE PREDICTORS
predictions = backtest(weather1, rr, predictors)
```

```python
# Calling the predictions
predictions
# The prediction below skips the first ten years as defined in the backtest() function
```

```python
# Checking for the accuracy of the prediction using accuracy metric for this case, I used Mean Absolute Error(MAE)
from sklearn.metrics import mean_absolute_error # importing mean_absolute_error from sklearn.metrics
mean_absolute_error(predictions["actual"], predictions["prediction"])
```

```python
# Alternative method of getting mean absolute error using mean() function
predictions["diff"].mean() 
# The value below means that the predictions are 5 degrees half of the actual temperatures which is a huge error and I had to improve on this.
```

```python
# Generating predictors to help in improving the accurracy by looking at the avarage temperature for the past few days e.g 3 or 11 days rather than focussing on the data for the previous day only. For example it sometimes rains in some days during summer thus biasness is reduced.
def pct_diff(old, new):
    return(new - old)/old
def compute_rolling(weather, horizon, col):
    label = f"rolling_{horizon}_{col}"
    
    weather1[label] = weather1[col].rolling(horizon).mean()
    
    weather1[f"{label}_pct"] = pct_diff(weather1[label], weather1[col])
    return weather1

rolling_horizons = [3, 14]
for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        weather1 = compute_rolling(weather1, horizon, col)
```

```python
# Calling weather1 dataset
weather1
```

```python
# Removing the first 14 rows with the missing values
weather1 = weather1.iloc[14:,:]
```

```python
# Viewing the dataset
weather1
```

```python
# Filling in any missing value
weather1 = weather1.fillna(0)
```

```python
# Getting the mean of the current and previous rows 
def expand_mean(df):
    return df.expanding(1).mean()

for col in ["tmax","tmin","prcp"]:
    weather1[f"month_avg_{col}"] = weather1[col].groupby(weather1.index.month, group_keys = False).apply(expand_mean)
    weather1[f"day_avg_{col}"] = weather1[col].groupby(weather1.index.day, group_keys = False).apply(expand_mean)
```
