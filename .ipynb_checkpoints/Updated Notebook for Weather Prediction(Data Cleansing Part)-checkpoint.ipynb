# Data Cleaning

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading weather1 dataset which is in comman separetad variables format while indexing "DATE" column of the data frame.
weather1 = pd.read_csv("weather1.csv", index_col="DATE")
```

```python
# calling weather1 dataset which is a historical data of weather conditions around JFK INTERNATIONAL AIRPORT, NY US
weather1
```

```python
# Examining the dataset
weather1.info()
```

```python
# examining the first 5 rows
weather1.head()
```

```python
# Checking the summary statistics of the dataset using describe() function
print(weather1.describe())
```

```python
# Calculating the null percentage while using pd.isnull.sum() to sum all the null values in each column then divided by the total number of rows to get the percentage of null values
null_pct = weather1.apply(pd.isnull).sum()/weather1.shape[0]
```

```python
# Calling the percentage of null values in each column
null_pct
```

```python
# Checking out columns which are valid in the weather1 dataset i.e. where the percentage of null values is less than 0.5
valid_columns = weather1.columns[null_pct < .05]
```

```python
# Calling the valid columns
valid_columns
```

```python
# Changing the dataset to include only the valid columns using copy() function
weather1 = weather1[valid_columns].copy()
```
